# Modern lattice layout

PyTao ships an opt-in replacement for the default `lat_layout` graph, called the **modern layout**.
It draws each element from a Python-side `ModernLayoutConfig` that classifies elements by `key`
and `name`, then renders the result with either backend (Matplotlib or Bokeh).

---
## Tao setup

````python
%config InlineBackend.figure_format = "retina"
````

````python
import os

os.environ["BOKEH_ALLOW_WS_ORIGIN"] = "*"
````

````python

import matplotlib.pyplot as plt

from pytao import Tao
from pytao.plotting import ModernLayoutConfig
````

````python
tao = Tao(init_file="$ACC_ROOT_DIR/bmad-doc/tao_examples/cbeta_ffag/tao.init", plot="bokeh")
````

## Enable the modern layout

Set `layout_style` on whichever graph manager(s) you want. Set it to `None` to revert.

````python
cfg = ModernLayoutConfig.default()

tao.bokeh.layout_style = cfg
tao.matplotlib.layout_style = cfg
````

To make the modern layout the default for every new manager:

```python
tao.bokeh.configure(layout_style=cfg)
tao.matplotlib.configure(layout_style=cfg)
```

## Bokeh: beta on top, modern layout below

````python
tao.plot("beta", include_layout=True, width=900, xlim=(0, 5), backend="bokeh")
````

## Matplotlib: same plot, same modern layout

````python
tao.plot("beta", include_layout=True, figsize=(9, 4), backend="mpl", xlim=(0, 5))
````

## BPM naming

Some lattices use the `MARKER` key for BPMs (e.g. `FA.BPM01`). By default, any `MARKER`
whose name contains `BPM` renders with the BPM style; other markers are skipped.

For a different convention, add a rule:

````python
from pytao.plotting import ClassifyRule

tao.bokeh.layout_style = cfg.with_overrides(
    rules=[
        *cfg.rules,
        ClassifyRule(key_pattern=r"^MARKER$", name_pattern=r"PROBE", style="BPM"),
    ]
)
````

## Customize colors and sizes

`with_overrides` returns a copy; `styles` accepts partial merges.

````python
highlighted = cfg.with_overrides(
    styles={
        "KICKER": {"color": "#FF1493", "height": 36},
        "QUADRUPOLE": {"color": "#1E90FF"},
    }
)

tao.bokeh.layout_style = highlighted
tao.plot("beta", include_layout=True, width=900, backend="bokeh", xlim=(0, 5))
````
